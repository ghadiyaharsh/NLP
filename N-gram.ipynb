# **Implement N-gram Language model**

### What is an N-gram Language Model?

It predicts the next word in a sequence using the previous 
N−1 words.

Unigram (N=1): Probability of a single word.

Bigram (N=2): Probability of word given the previous word.

Trigram (N=3): Probability of word given previous 2 words.

```python
import nltk
from nltk.util import ngrams
from collections import Counter,defaultdict
```

```python
text = "I love natural language processing and I love machine learning"
```

```python
#step-1 tokenization 
tokens = nltk.word_tokenize(text.lower())
#Breaks text into words.
#Example: "I love natural language processing and I love machine learning" → ["i", "love", "natural", "language", "processing", "and", "i", "love", "machine", "learning"].
```

```python
bigrams = list(ngrams(tokens,2))
#creates all consecutive word pairs (bigrams).
#Example: ["i", "love", "natural", "language", "processing", "and", "i", "love", "machine", "learning"] → [("i", "love"), ("love", "natural"), ("natural", "language"), ("language", "processing"), ("processing", "and"), ("and", "i"), ("i", "love"), ("love", "machine"), ("machine", "learning")]
```

```python
bigram_counts = Counter(bigrams)
unigram_counts = Counter(tokens)

'''Counts how many times each pair (bigram) appears.
Counts how many times each word (unigram) appears.
Example: bigram_counts[("i", "love")] = 2, unigram_counts["i"] = 2

'''
```

```python
def bigram_probability(w1, w2):
    return bigram_counts[(w1, w2)]/unigram_counts[w1] 

#probability of seeing word2 after word1
```

```python
print("P(language | natural) =", bigram_probability("natural", "language"))
print("P(love | I) =", bigram_probability("i", "love"))
```
